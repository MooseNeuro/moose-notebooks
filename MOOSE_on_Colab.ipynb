# Released binary package on Google Colab
You can install released binaries of MOOSE on a Colab notebook environment using `pip`:
```
pip install pymoose
```
will install the latest released version from PyPI.

# Using development versions of MOOSE on Google Colab
Alternatively, if you want to use the latest development version of MOOSE, you can directly build and install MOOSE from github repository on a Colab notebook environment.

1. First, install the dependencies. Colab uses `apt` as the package manager.

````python
# Install dependencies for building MOOSE
!sudo apt install libgsl-dev libhdf5-dev
````

2. Next, use `pip` to install the desired branch from the github repo. The format is `pip install git+https://{address}`.
Here {address} should be replaced by the web address of the repo. By default it installs the `main` branch (called `master` in older perlance).

  For example, `pip install git+https://github.com/subhacom/moose-core` will install master branch `subhacom`'s copy (fork) of the moose source code.
To install a specific branch, you have to appen `@{branchname}` to address.

  Thus `pip install git+https://github.com/subhacom/moose-core@fix495merge` will install the `fix495merge` branch from `subhacom`'s copy of the repo.

  You may see this message at the end, but ignore it:
  ```
  ERROR: pip's dependency resolver does not currently take into account all the packages that are installed. This behaviour is the source of the following dependency conflicts.
google-colab 1.0.0 requires ipykernel==5.5.6, but you have ipykernel 6.29.5 which is incompatible.
  ```

````python
!pip install git+https://github.com/subhacom/moose-core@fix495merge
````
